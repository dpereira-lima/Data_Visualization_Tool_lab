```python
import pandas as pd

# Carregar o dataset para visualização inicial
data = pd.read_csv('https://query.data.world/s/wqz2leixvgliwedd345tf54ffzcbtt?dws=00000')

# Mostrar as primeiras linhas do arquivo para análise
data.head()
```

```python
# Definindo as colunas que serão mantidas e as que serão transformadas
columns_to_keep = ['Category', 'Region']
columns_to_transform = data.columns.difference(columns_to_keep)

# Transformar as colunas C a I (todas as outras) em duas colunas: 'Description' e 'Value'
transformed_data = data.melt(id_vars=columns_to_keep, 
                             value_vars=columns_to_transform, 
                             var_name='Description', 
                             value_name='Value')

# Visualizar o resultado
transformed_data.head()
```

```python
# Exportar o dataframe para um arquivo CSV
transformed_data.to_csv('Out-of-School_Rates_and_Numbers_updated-2.csv', index=False)
```
