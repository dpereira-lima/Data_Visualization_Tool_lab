```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#carregar dados
df = pd.read_csv('https://query.data.world/s/dh3stfzlxrwyv5z3nq4smjmuiuytov?dws=00000')
```

```python
#analisando primeiras linhas do dataset
df.head()
```

```python
#explorando estrutura do dataset
df.info()
```

```python
#grafico de barras

#ajuste grfico
colors = ['red' if value < 0 else 'skyblue' for value in df['December Average Percent Increase']]

#configurando o grfico de barras
plt.figure(figsize = (10, 6))
plt.barh(df['Purchase Category'], df['December Average Percent Increase'], color = colors)

#adcinando ttulos e rtulos
plt.title('How Does Our Spending Change in December?', fontsize = 14)
plt.tight_layout()

#apresentar grfico
plt.savefig('Python chart.jpg', format = 'jpg', dpi = 300, bbox_inches = 'tight')
plt.show()
```
