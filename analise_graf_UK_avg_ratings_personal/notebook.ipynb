```python
import pandas as pd
```

```python
df = pd.read_csv('UK Average Ratings of Personal Well-being 2012-2023.csv')
```

```python
df.head()
```

```python
# Função para converter o período em duas datas no formato dd/mm/yyyy
def split_time_period(period):
    # Separar os períodos com base no "to"
    start_period, end_period = period.split(' to ')
    
    # Converter o mês e ano do início e do fim para o formato desejado
    start_date = pd.to_datetime(start_period + ' 01', format='%B %Y %d').strftime('%d/%m/%Y')
    
    # Para o final do período, pegar o último dia do mês
    end_date = pd.to_datetime(end_period + ' 01', format='%B %Y %d') + pd.offsets.MonthEnd(0)
    end_date = end_date.strftime('%d/%m/%Y')
    
    return start_date, end_date
```

```python
df[['Start data', 'End data']] = df['Time Period'].apply(lambda x: pd.Series(split_time_period(x)))
```

```python
df.to_csv('UK Average Ratings of Personal Well-being 2012-2023 - 2.csv', index=False)
```
